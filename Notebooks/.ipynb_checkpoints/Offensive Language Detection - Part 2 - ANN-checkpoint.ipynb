- Important params
     - Units: We tell it how many units we want to have in the first hidden layer  
     - activation function: These fxns are used to process the values from neurons in one layer before they are fed into the next layer. They can be relu, sigmoid, hyperbolic tangents etc 
     - input_dim: The number of features in the data
 
- Compile the model
- Important Params:
    - Optimizer: This is an algorithm which defines how the neural network updates the weights & learning rate in order to minimize the loss
    - Loss: function that the neural network & the optimizer use to compare predicted values to target values during training. different loss fns for regression and classification tasks.

## A. Training ANN with Imbalanced Data (Original Data) + Class weights

```python
bow = pd.read_csv('bow_train_im.zip', compression = 'zip').iloc[:,1:]
w2v = pd.read_csv('w2v_train_im.zip', compression = 'zip').iloc[:,1:]
glove = pd.read_csv('glove_train_im.zip', compression = 'zip').iloc[:,1:]
```

**Obtaining Class Weights**

```python
neg =  bow['label'].value_counts()[0]
pos =  bow['label'].value_counts()[1]
total = bow['label'].shape[0]
weight_for_0 = (1/neg) * (total/2)
weight_for_1 = (1/pos) * (total/2)

#create a dictionary to hold the weights
class_weights = {0:weight_for_0, 1:weight_for_1 }
class_weights
```

**Setting model parameters** 

```python
#create the callback to reduce the learning rate 
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=0.001)

#get the correct value for the initial bias
initial_bias = np.log([pos/neg])
output_bias = tf.keras.initializers.Constant(initial_bias)

#Define the model metrics 
metrics=[tf.keras.metrics.BinaryAccuracy(name='accuracy')]
```

**Function to convert predictions from probablity to class**

```python
#create a function that will take the prediction treshold and convert the values and to convert predictions to pd series

def convert(prediction_array, threshold):
    predictions_list = []
    for val in prediction_array:
        for value in val:
            if value >= threshold:
                predictions_list.append(1)
            else:
                predictions_list.append(0)
    return pd.Series(predictions_list)
    
```

**Function to plot Learning curves**

```python
#create a function to plot training & validtion metrics

def train_plot(fitted_model):

    plt.figure( figsize=(20,20))

    plt.subplot(2, 2, 1)
    plt.plot(fitted_model.history['val_loss'])
    plt.plot(fitted_model.history['loss'])
    plt.title('Loss - Training vs Testing')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['val_loss','loss'], loc='upper left')

    plt.subplot(2, 2, 2)
    plt.plot(fitted_model.history['val_accuracy'])
    plt.plot(fitted_model.history['accuracy'])
    plt.title('Accuracy - Training vs Testing')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['val_accuracy','accuracy'], loc='upper left')
```

**Function to collate test results**

```python
#create function to print collate test results 

def test_results(true_labels, predicted_labels,model_name, dataset):
    
    f1score = f1_score(true_labels, predicted_labels, average = 'macro')
    cm = confusion_matrix(true_labels, predicted_labels)
    per_class = cm.diagonal()/cm.sum(axis=1)
    result = pd.Series({'Model Name': model_name , 'Dataset':dataset,'Macro_F1score':f1score, 'Majority_Class_Accuracy': per_class[0], 'Minority_Class_Accuracy': per_class[1]})
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
    disp.plot()
    plt.show()
    return result
```

**Define function for creating the ANN**

```python
#Define a function that creates a neural network with two hidden layers and sigmoid output layer
def create_model(input_shape,metrics=metrics, output_bias=None):
    
    inputs = layers.Input(shape = input_shape, name = 'Input Layer')
    hidden_1 = layers.Dense(units = 128, activation = 'relu', name ='hidden_layer_1')
    hidden_2 = layers.Dense(units = 128, activation = 'relu', name ='hidden_layer_2')
    outputs = layers.Dense(units = 1, activation = 'sigmoid', name = 'predictions', bias_initializer=output_bias)
    
    model = Sequential([inputs, hidden_1, hidden_2,outputs])
    model.compile(optimizer=tf.keras.optimizers.Adam(),loss=tf.keras.losses.BinaryCrossentropy(), metrics=metrics)
    
    return model
```

### Training ANN with Imbalanced BOW

**Notes on Outputs of the BERT encoder**
- Pooled output: This represents the embeddings for each input (row in dataset, tweet etc) as a whole. so its like the sentence vector for that row. For distilbert, this will be (1,768) since distilbert outputs 768 hidden states for each token ad the hidden states for each token in the sentence will then be aggregated.
- Sequence output: This is like a tensor which has the hidden states for each token in the sentence. size will be 1 x 128 x 768. Since this distilbert model gives 128 (with CLS,token,SEP,padding) for each sentence.

For fine tuning the bert model for classification, I will be using the pooled outputs
